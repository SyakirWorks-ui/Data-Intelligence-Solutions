```python
# Load hasil prediksi untuk divalidasi
output_file = os.path.abspath(os.path.join('..', 'data', 'processed', 'final_predictions.csv'))

if os.path.exists(output_file):
    df_results = pd.read_csv(output_file, sep=';')
    
    print("--- Ringkasan Hasil Prediksi Otomatis ---")
    print(f"Total Pelanggan Diproses: {len(df_results)}")
    print(f"Prediksi Churn (1): {df_results['Churn_Prediction'].sum()} orang")
    print(f"Prediksi Stay  (0): {(df_results['Churn_Prediction'] == 0).sum()} orang")
    
    print("\n--- 5 Sampel Prediksi Tertinggi (Probabilitas Churn) ---")
    display(df_results.sort_values(by='Churn_Probability', ascending=False).head())
else:
    print("‚ùå File hasil prediksi belum tersedia. Pastikan Pipeline berjalan sukses.")
```
